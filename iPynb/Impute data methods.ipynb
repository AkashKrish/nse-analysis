# Data Imputing

```python
import pandas as pd
from scipy.stats import mode
import numpy as np
```

```python
#Laod Data File
input_file_name='C:\\users\\Akash\\Documents\\Github\\PyThon\\Data Sets\\ACCS\\Kudwi\\Kudwi.csv'
df = pd.read_csv(input_file_name)
df['Date'] = pd.to_datetime(df['Start_Time'], format='%d-%m-%Y %H:%M').dt.date
df['Time'] = pd.to_datetime(df['Start_Time'], format='%d-%m-%Y %H:%M').dt.time

column_order = ['Salon Code', 'Date', 'Time', 'Customer ID',
                'ProdID', 'EmpID', 'Qty', 'Tax', 'Retail_Price',
                'Discount_Amt',	'Discount_Perct', 'Total',
                'Mem_Disc', 'Oth_Disc', 'Studio_Status']
df = df[column_order]
```

### Find number of missing columns

```python
def num_missing(x):
  return sum(x.isnull())
df.apply(num_missing, axis=0) #axis = 1 will gib=ve number of columns empty per row
```

## Imputing using mode from scipy

```python
df['Customer ID'].fillna(mode(df['Customer ID']).mode[0], inplace=True)
```

## Pivot table

```python
df.pivot_table(values=['Total'], index=['Customer ID', 'EmpID', 'ProdID'], aggfunc=np.sum)
```

### Use crosstab for initial hypothesis testing

Initially assumed that store A had more business

```python
pd.crosstab(df['Salon Code'], df['Total'], margins=True)
```
